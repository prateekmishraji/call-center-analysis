**Most common Reason for Calls**

```python
df['reason'] = df['title'].apply(lambda value:value.split(":")[0])
```

**Using seaborn to create a countplot of 911 calls by Reason**

```python
sns.countplot(x='reason',data=df,palette='viridis')
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
df['hour'] = df['timeStamp'].apply(lambda time : time.hour)
df['month'] = df['timeStamp'].apply(lambda time : time.month)
df['dayofweek'] = df['timeStamp'].apply(lambda time : time.dayofweek)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
dmap
```

```python
df['dayofweek'] = df['dayofweek'].map(dmap)
```

**create a countplot of the Day of Week column with the hue based off of the Reason column**

```python
sns.countplot(x="dayofweek",data=df,palette='viridis',hue='reason')
```
